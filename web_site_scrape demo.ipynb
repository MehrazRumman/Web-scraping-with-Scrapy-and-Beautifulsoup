```python
from urllib.request import urlopen,Request
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
s='https://getlasso.co/affiliate-programs/'
hdr={'User-Agent': 'Mozilla/5.0'}
req=Request(s,headers=hdr)
page=urlopen(req)
bs = BeautifulSoup(page)
all_link=bs.find_all('div',class_='col-2 text-right')
cnt=0
links=[]
groups=[]
for link in all_link:
    t=link.find('a')
    cnt=cnt+1
    links.append('https://getlasso.co'+t['href'])
```

```python
for j in all_link:
    tt=j.find('a')
    tt=tt['href']
    tt=tt.replace('/niche/','')
    tt=tt.replace('/','')
    tt='Best '+tt+' Affiliate Programs'
    groups.append(tt)

```

```python
k=0
stract={
    'groups':[],
    'title':[],
    'Description':[],
    'Commission_Rate':[],
    'Cookie_duration':[]
    
}
cnt1=0
for i in links:
    page_unit=i
    req_page=Request(page_unit,headers=hdr)
    unit=urlopen(req_page)
    bs_unit=BeautifulSoup(unit)
    title_pick=bs_unit.find('div',class_='lasso-box-2')
    if title_pick is not None:
        pick_t=title_pick.find('a',class_='lasso-title').text
        pick_des=title_pick.find('div',class_='lasso-description')
        pick_d= pick_des.find_all('p')
        l=[]
        for pick in pick_d:

            a=pick.text
            a=a.replace('\n','')
            a=a.replace('Commission Rate:','')
            a=a.replace('Cookie Duration:','')
            l.append(a)
            cnt1=cnt1+1
        stract.setdefault('groups',[]).append(groups[k]) 
        stract.setdefault('title',[]).append(pick_t)
        stract.setdefault('Description',[]).append(l[0])
        stract.setdefault('Commission_Rate',[]).append(l[1])
        stract.setdefault('Cookie_duration',[]).append(l[2])
      
       # stract.update({'groups':[].append(groups[k]),'title':[].append(pick_t),'Description':[].append(l[0]),'Commission_Rate':[].append(l[1]),'Cookie_duration':[].append(l[2])})

    other_pro=bs_unit.find_all('li',class_='h2')
    if other_pro is not None:
    
        for o in other_pro:
            #if o.find('h2',class_='mb-3') is not None:
            tit=o.find('h2',class_='mb-3')
            if tit is not None:
                tit=tit.text
                
                d=o.find_all('p',class_='mb-3')
                l1=[]
                for g in d:
                    gg=g.text
                    gg
                    gg=gg.replace('Commission Rate:','')
                    gg=gg.replace('Cookie Duration:','')
                    l1.append(gg)
                if len(l1)!=0:
                    stract.setdefault('groups',[]).append(groups[k]) 
                    stract.setdefault('title',[]).append(tit)
                    stract.setdefault('Description',[]).append(l1[0])
                    stract.setdefault('Commission_Rate',[]).append(l1[1])
                    stract.setdefault('Cookie_duration',[]).append(l1[2])
                
    k=k+1
   
```

```python

df=pd.DataFrame(stract)
df.head()
```
